```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('miles.csv')
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize = (12,4))
plt.plot(df.miles)
plt.show()
```

```python
sns.histplot(df.miles, kde = True)
plt.show()
```

```python
mean_population = df.miles.mean()
std_population = df.miles.std()

mean_population, std_population 
```

```python
SAMPLE_SIZE = 50

sample = df.sample(SAMPLE_SIZE, random_state = 100)
sample.head()
```

```python
sample.shape
```

```python
mean_sample = sample.miles.mean()
std_sample = sample.miles.std()
mean_sample, std_sample 
```

```python
import scipy.stats as st
import numpy as np

z_score_95_confidence = st.norm.ppf(0.975)
z_score_95_confidence
```

```python
margin_of_error = z_score_95_confidence * (std_population/ np.sqrt(SAMPLE_SIZE))
margin_of_error
```

```python
lower = mean_sample - margin_of_error
upper = mean_sample + margin_of_error
lower, upper 
```
